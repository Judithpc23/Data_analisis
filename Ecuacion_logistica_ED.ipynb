## **Modelo de la ecuación Logistica**

```python
# Definiendo los parámetros
P0 = 6104.3955855705235
r = 0.0371445640864125
K = 12066.92748462337

# Definiendo la función logística
def logistic(t):
    return (K * P0 * np.exp(r * t)) / (K + P0 * (np.exp(r * t) - 1))

# Creando una serie de valores de tiempo t para graficar
t = np.linspace(0, 10, 20) # Desde t=0 hasta t=50 con 500 puntos

# Calculando la población P(t) para cada valor de t
P = logistic(t)
print(t)
print(P)

# Graficando
plt.figure(figsize=(10,6))
plt.plot(t, P, label='P(t)')
plt.xlabel('Tiempo(Años)')
plt.ylabel('Cantidad de estudiantes becados')
plt.title('Crecimiento de la porblacion de becados modelado con la ecuación logística')
plt.legend()
plt.grid(True)
plt.show()

df = pd.DataFrame({'Tiempo (Años)': t, 'Cantidad de Estudiantes Becados': P})
df
```

### **Funcion que establece la estructura del modelo**

```python
# Definir la función de la ecuación logística
def logistic_growth(t, P0, r, K):
    return (K * P0 * np.exp(r * t)) / (K + P0 * (np.exp(r * t) - 1))
```

### **MODELO CON AJUSTE DE LA CURVA**

```python
# Datos
datos = [
    [6018, 6395],
    [6325, 6281],
    [6723, 6408],
    [6965, 6686],
    [7296, 6956],
    [7830, 7796],
    [9759, 9997],
    [10225, 10001],
    [10223, 9805],
    [9803, 9513],
    [9342, 9595],
    [9401, 8860],
    [9129, 8529]
]

fila = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
columna = ['-1', '-2']

# Crear una lista de años y semestres combinados
años_semestres = []
semestre = ["-1","-2"]
for año, semestres in zip(fila, datos):
    for semestre, poblacion in zip(columna, semestres):
        años_semestres.append(f"{año}{semestre}")

# Poblaciones
poblaciones = np.array([poblacion for semestres in datos for poblacion in semestres])

# Años para el ajuste
años_ajuste = np.arange(len(años_semestres))

# Ajustar el modelo logístico
popt, pcov = curve_fit(logistic_growth, años_ajuste, poblaciones, p0=[poblaciones[0], 0.1, max(poblaciones)*1.5])

# Imprimir los parámetros estimados
P0_est, r_est, K_est = popt
print("P0 =", P0_est, "r =", r_est, "K =", K_est)

# Gráfica de los datos y el modelo ajustado
plt.figure(figsize=(12, 6))
plt.scatter(años_ajuste, poblaciones, color='red', label='Datos reales')
plt.plot(años_ajuste, logistic_growth(años_ajuste, *popt), label='Ajuste logístico', color='blue')
plt.xticks(ticks=años_ajuste, labels=años_semestres, rotation=45)
plt.xlabel('Año y Semestre')
plt.ylabel('Población')
plt.title('Ajuste de la ecuación logística a los datos de población por año y semestre')
plt.legend()
plt.show()

tabla_valores = pd.DataFrame({
    'Año y Semestre': años_semestres,
    'Población Real': poblaciones,
    'Población Ajustada': logistic_growth(años_ajuste, *popt)
})

# Mostrar la tabla
tabla_valores
```

```python
tabla_valores_error = pd.DataFrame(
     {
    'Año y Semestre': años_semestres,
    'Población Real': poblaciones,
    'Población Ajustada': logistic_growth(años_ajuste, *popt),
    'Error Absoluto' : abs(poblaciones - logistic_growth(años_ajuste, *popt)),
    'Error Relativo' : ((abs(poblaciones - logistic_growth(años_ajuste, *popt)))/abs(poblaciones)),
    '% Error Relativo' : ((abs(poblaciones - logistic_growth(años_ajuste, *popt)))/abs(poblaciones))*100
})

tabla_valores_error
```
